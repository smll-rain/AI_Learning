```python
age_train_subset_data.shape
```

##### 标签构造

```python
# 标签构造
def sex_transfomer(value):
    if value == 1:
        return 1
    else:
        return 0
age_train_lable = age_train_subset_data.attr_sex_female.apply(sex_transfomer)
age_train_lable.head()
```

```python
age_train_lable.shape
```

##### 剔除无关特征

```python
# 训练集删除无用的列及方差为0的特征
age_train_subset_data.drop(columns=['attr_sex_male', 'attr_sex_female','attr_sex_other','attr_level_4', 'attr_level_12'], axis=1,inplace=True)
age_train_subset_data.head()
```

```python
# 测试集删除无用特征
age_prediction_data.drop(columns=['attr_sex_male', 'attr_sex_female','attr_sex_other','attr_level_4', 'attr_level_12'], axis=1,inplace=True)
age_prediction_data.head()
```

##### 训练集随机采样

```python
# 数据集打散
from sklearn.model_selection import train_test_split
age_train_X,age_test_X, age_train_y, age_test_y =train_test_split(age_train_subset_data, age_train_lable, test_size=0.3, random_state=42)
print(age_train_X.shape)
print(age_train_y.shape)
print(age_test_X.shape)
print(age_test_y.shape)
```

##### 随机采样数据平衡性检查

##### 参数调优

```python
%%time
# 参数调优
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import  GridSearchCV
C = np.arange(0.1,2, 0.1)
params = {'C':C}
age_gs = GridSearchCV(LogisticRegression(random_state=42),param_grid=params, cv=5)
age_gs.fit(age_train_X, age_train_y)
print(age_gs.best_score_)
print(age_gs.best_params_)
```

##### 模型训练及预测

```python
%%time
lr_model = LogisticRegression(C=0.8,solver= 'lbfgs')
lr_model.fit(age_train_X, age_train_y)
print('训练集准确度：%f' % lr_model.score(age_train_X, age_train_y))
print('测试集准确度：%f' % lr_model.score(age_test_X, age_test_y))
```

#### 递归选择特征

```python
%%time
from sklearn.ensemble import AdaBoostClassifier
# 模型训练
enhance_feature_dict = {}
weak_feature_dict = {}
unchange_feature_dict = {}
original_train_score = 0.959680
original_test_score = 0.960143
feature_len = len(cleaned_train_X.columns)
feature_index = 1
for feature in cleaned_train_X.columns:
    rc_train_X = train_X.drop(columns=[feature], axis=1)
    rc_test_X = test_X.drop(columns=[feature], axis=1)
    rc_model = AdaBoostClassifier(learning_rate=0.9, n_estimators=350)
    rc_model.fit(rc_train_X, train_y)
    rc_train_score = rc_model.score(rc_train_X, train_y)
    rc_test_score = rc_model.score(rc_test_X, test_y)
    if rc_train_score > original_train_score and rc_test_score > original_test_score:
        enhance_feature_dict[feature] = [rc_train_score, rc_test_score]
    elif rc_train_score == original_train_score and rc_test_score == original_test_score:
        unchange_feature_dict[feature] = [rc_train_score, rc_test_score]
    else:
        weak_feature_dict[feature] = [rc_train_score, rc_test_score]
    print('\r 筛选进度：%0.2f%%' % (feature_index / feature_len * 100), end='')
    feature_index += 1
print('\n')
print('递归删除后对模型效果有增强作用的特征：\n' , enhance_feature_dict)
print('递归删除后对模型效果有减弱作用的特征：\n' , weak_feature_dict)
print('递归删除后对模型效果有不变作用的特征：\n' , unchange_feature_dict)
```

**递归单特征筛选结果**

- 特征attr_level_7的删除对模型产生了积极的效果，这说明该特征对模型起到了一定的干扰作用，即为噪声
- 同时需要留意，虽然对模型来说是噪声，但不代表该特征与其它特征之间不存在关联性，需要结合活动背景、时间段等进一步分析确认
- 删除特征attr_level_7后的增强效果，比逻辑回归模型筛选的特征效果略低一筹。

### 总结

#### 存在的问题
**脱敏的问题**
- 优点
  - 脱敏有利于保护数据的安全
  - 结构化数据 ，减少了特征工程耗费的时间(但也给分析挖掘带来了难度)
- 缺点
  - 数据分布变得更加零散，需要还原数据才能比较好的把握数据的分布
  - 很多属性的具体含义难以捉摸
  - 有些产品可能会有自身的特性，这种特性会因脱敏而被隐藏
  - 0/1二值化处理，将一些连续性的属性变得很零散
  
**数据样本问题**
- 连续性的数据存在断层，如年龄、促销敏感度等，这些都是具有连续性的属性，但样本只有其中的一部分，可能容易造成模型的偏好
- 分析时，因数据仅取其中一小部分，对年龄、促销敏感度整个区间的划分不清晰，导致无法把握全局，容易产生主观上的分析偏差
- 加重冷启动问题。因年龄、用户等级、促销敏感度等未给全信息，即信息缺失，会导致模型对新增数据的兼容性较差


#### 模型总结
**时间优先，可以牺牲一点精度，优选逻辑回归**
- 逻辑回归，预测速度比随机森林快2.42倍，比AdaBoost快10.47倍，比GBDT快39.18倍，比XGBoost快42.64倍。

**精度优先，时间充足，优选GBDT**
- GBDT，预测精度比逻辑回归高出1.26个百分点，比随机森林高出0.29个百分点，比AdaBoost高出0.24个百分点，比XGBoost高出0.006个百分点。

**时间精度两者都需要兼顾，优选AdaBoost**
- AdaBoost，时间比逻辑回归慢10.47倍左右，但精度比逻辑回归高出1.02个百分点；比GBDT的精度低0.24个百分点，但时间要节省2.5倍。

**模型改进效果**
- 通过随机逻辑回归提取特征后的模型，精确度提升了0.01%，同时也过滤很多特征，运行速度提升了21.13%，有明显的改进效果，但因特征过滤较多，需进步研究其业务的可解释性问题
