### 数据预处理

#### 数据合并

```python
col_name = test_data.columns.tolist()
col_name.insert(1,'Survived')
test_data_reindex = test_data.reindex(columns=col_name)
concat_data = pd.concat([train_data,test_data_reindex])
print(concat_data.head())
print(concat_data.tail())
```

#### 缺失值填充

方式：

    平均值
    经验值
    回归模型预测
    剔除，一般大于30%左右可以考虑剔除掉
    

Age缺失值用平均值填充

Embarked缺失值用众数S填充

训练集与预测集合并一起填充缺失值

```python
import matplotlib.pyplot as plt
import seaborn as sns
f, ax = plt.subplots(1,2, figsize=(15, 8))
sns.violinplot('Pclass', 'Age', hue='Survived', data=train_data, split=True, ax=ax[0])
sns.violinplot('Sex', 'Age', hue='Survived', data=train_data, split=True, ax=ax[1])
ax[0].set_title('Pclass & Age & Survived')
ax[1].set_title('Sex & Age & Survived')
plt.show()
```

年龄在20-50之间，幸存的人数较多

20-30岁之间的幸存人数最多

在处理特征的时候，发现Name存在着年龄区分的称呼，如Mr、 Mrs、Miss等，可以用此规律对年龄进行有针对性的填充

```python
name_list = concat_data.Name.apply(lambda x : x.split(',')[1].split('.')[0].strip())
concat_data['Surname'] = name_list
```

```python
surname_mean = concat_data[['Age', 'Surname']].groupby('Surname').mean()
print(surname_mean)
```

```python
for index, value in zip(surname_mean.index,surname_mean.values):
    concat_data.loc[(concat_data.Age.isnull())&(concat_data.Surname==index), 'Age']=value[0]
```

```python
concat_data.groupby('Surname').Age.mean()
```

```python
concat_data.Embarked.fillna(value="S", inplace=True)
```

#### 数据二值化

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
one_hot_data = pd.get_dummies(concat_data[['Sex','Embarked']])
cleaned_data = pd.concat([concat_data, one_hot_data], axis=1)
```

### 特征工程

#### 年龄离散化

```python
# 0-80分成5个箱子，间距16
cleaned_data['Age_bin'] = 0
cleaned_data.loc[cleaned_data.Age<=16, 'Age_bin'] = 0
cleaned_data.loc[(cleaned_data.Age>16) & (cleaned_data.Age<=32), 'Age_bin'] = 1
cleaned_data.loc[(cleaned_data.Age>32) & (cleaned_data.Age<=48), 'Age_bin'] = 2
cleaned_data.loc[(cleaned_data.Age>48) & (cleaned_data.Age<=64), 'Age_bin'] = 3
cleaned_data.loc[(cleaned_data.Age>64) & (cleaned_data.Age<=80), 'Age_bin'] = 4
cleaned_data.head()
```

#### k折交叉验证

```python
from sklearn.model_selection import KFold, cross_val_score, cross_val_predict
kfold = KFold(n_splits=10, random_state=22)
cross_mean_list = []
accuracy_list = []
cross_std_list = []
y_pred = []
classifiers = ['Linear Svm', 'Radial Svm', 'Logistic Regression',
              'KNN', 'DecisionTree','GaussianNB', 'Random Forest']
models = [svm.SVC(kernel='linear'), svm.SVC(gamma='auto',kernel='rbf'), LogisticRegression(solver='lbfgs'), 
          KNeighborsClassifier(), DecisionTreeClassifier(),GaussianNB(), 
          RandomForestClassifier(n_estimators=10)]
for model in models:
    cross_val_result = cross_val_score(model,train_X,train_y, cv=kfold, scoring='accuracy')
    cross_mean_list.append(cross_val_result.mean())
    cross_std_list.append(cross_val_result.std())
    accuracy_list.append(cross_val_result)
    y_pred.append(cross_val_predict(model,train_X,train_y, cv=kfold))
model_result_data = pd.DataFrame({'cross_mean':cross_mean_list, 'cross_std':cross_std_list}, index=classifiers)
print(model_result_data)
```
